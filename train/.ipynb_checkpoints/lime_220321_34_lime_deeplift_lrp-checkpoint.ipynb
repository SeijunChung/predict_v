```python
import numpy as np
import tensorflow as tf

class Conv1dLayers():
    def __init__(self,
                 name='cnn',
                 batch_size=100,
                 input_shape=(),
                 filters=(),
                 kernel_sizes=(),
                 kernel_initializer='glorot_uniform',
                 strides=(),
                 pool_sizes=(),
                 drop_outs=(),
                 activations=(),
                 batch_normal=(),
                 batch_norm_pars=(),
                 conv_names=(),
                 y_concat_ch=0):
        self.name=name
        self.batch_size=batch_size
        self.input_shape=input_shape
        self.output_shape=(0,)
        self.filters=filters
        self.kernel_sizes=kernel_sizes
        self.convs=[]
        self.conv_names=[]
        self.y_concat_ch=y_concat_ch
        self.activation=[]
        self.batch_normal=[]
        self.pools=[]
        self.dropouts=[]
        self.conv_output_sizes=[]
        self.pool_output_sizes=[]
        self.layer_outputs = []
        
        conv_function=tf.keras.layers.Conv1D
        conv_name_sub='_cv1d_l'
        
        ulay_inp_size = input_shape[0]
        ft_output_shape = 0
        for i,(ft,kn,st) in enumerate(zip(filters,kernel_sizes,strides)):
            ft_output_shape = ft
            conv_name=conv_names[i] if len(conv_names)>i and len(conv_names[i])>0 else self.name+conv_name_sub+str(i+1)
            pool_name=self.name+'_pool_l'+str(i+1)
            drop_name=self.name+'_drop_l'+str(i+1)
            
            # Check activation function name
            act = activations[i] if len(activations)>i and type(activations[i]) is type(tf.nn.leaky_relu) else None
            self.activation.append(act)
            
            # Check batch_normal
            if len(batch_normal) > i and batch_normal[i]==True:
                m,e=0.99,0.001
                if len(batch_norm_pars)>1:
                    if batch_norm_pars[0]>0 and batch_norm_pars[0]<1.:
                        m=batch_norm_pars[0]
                    if batch_norm_pars[1]>0 and batch_norm_pars[1]<1.:
                        e=batch_norm_pars[1]
                self.batch_normal.append(tf.keras.layers.BatchNormalization(momentum=m, epsilon=e, name=self.name+conv_name_sub+str(i+1)+'_bn'))
            else:
                self.batch_normal.append(None)
            
            # define conv
            self.convs.append(conv_function(
                filters=ft, kernel_size=kn,strides=st,
                kernel_initializer=kernel_initializer,
                padding='same',
                input_shape=input_shape,name=conv_name))
            self.conv_names.append(conv_name)
            
            # output size of conv
            self.conv_output_sizes.append(np.ceil(ulay_inp_size/st))
            
            # Check pool size for each layer: if pool size is small don't do max pool
            po = pool_sizes[i] if len(pool_sizes) > i else 0
            if po>1:
                po=int(po)
                self.pools.append(tf.keras.layers.AveragePooling1D(pool_size=po,strides=2,padding='same',name=pool_name))
                self.pool_output_sizes.append(np.ceil(self.conv_output_sizes[i]/po))
            elif po<0:
                po=-int(po)
                self.pools.append(tf.keras.layers.MaxPool1D(pool_size=po,strides=2,padding='same',name=pool_name))
                self.pool_output_sizes.append(np.ceil(self.conv_output_sizes[i]/po))
            else:
                self.pools.append([])
                self.pool_output_sizes.append(self.conv_output_sizes[i])
            
            ulay_inp_size = self.pool_output_sizes[i]
            
            # Check drop out value for each layer: if necessary (do>=1.), don't drop out
            do = drop_outs[i] if len(drop_outs) > i else 0
            if do>0. and do<1.:
                self.dropouts.append(tf.keras.layers.Dropout(do,name=drop_name))
            else:
                self.dropouts.append([])
        
        self.output_shape = (self.pool_output_sizes[-1], ft_output_shape)
    
    def __call__(self, input_tensor, is_training=False, y_concat=None):
        with tf.compat.v1.variable_scope(self.name,reuse=tf.compat.v1.AUTO_REUSE):
            x=input_tensor
            
            # layers for convs
            for conv, pool, dropout, act, bn in zip (self.convs, self.pools, self.dropouts, self.activation, self.batch_normal):
                # Concatenate conditioning vector on feature map axis.
                
                if y_concat!=None and self.y_concat_ch==y_concat.get_shape()[-1]:
                    x_shapes=x.get_shape()
                    x=tf.concat([x, tf.reshape(y_concat, [self.batch_size, 1, self.y_concat_ch])*
                                    tf.ones([self.batch_size, x_shapes[1], self.y_concat_ch])],-1)
                # 1d convolution
                x=conv(inputs=x)
                # batch normalization
                if bn is not None:
                    x=bn(x)
                # activation function
                tf.add_to_collection('DL', x)
                tf.add_to_collection('DL_ref', x)
                tf.add_to_collection('STD', x)
                if act is not None:
                    x=act(x)
                # pooling
                if pool!=[]:
                    x=pool(inputs=x)
                # dropout
                if dropout!=[]:
                    x = tf.compat.v1.where(is_training, dropout(inputs=x), x)
                self.layer_outputs.append(x)
                tf.add_to_collection('DL', x)
                tf.add_to_collection('DL_ref', x)
                tf.add_to_collection('STD', x)
            return x
            
    def flatten(self, input_tensor, is_training=False, y_concat=None):
        return tf.keras.layers.Flatten()(self.__call__(input_tensor, is_training, y_concat))
    
    def get_output_number(self):
    # final output number if flatten
        return np.prod(self.output_shape)
```

```python
import numpy as np
import tensorflow as tf
    
class Conv1dTransLayers():
    def __init__(self,
                 name='decnn',
                 batch_size=100,
                 input_shape=(),
                 output_shape=(),
                 filters=(),
                 kernel_sizes=(),
                 kernel_initializer='glorot_uniform',
                 strides=(),
                 pool_sizes=(),
                 drop_outs=(),
                 activations=(),
                 batch_normal=(),
                 batch_norm_pars=(),
                 conv_names=(),
                 y_concat_ch=0):
        self.name=name
        self.batch_size=batch_size
        self.input_shape=input_shape
        self.output_shape=output_shape
        self.filters=filters
        self.kernel_sizes=kernel_sizes
        self.convs=[]
        self.conv_names=[]
        self.y_concat_ch=y_concat_ch
        self.activation=[]
        self.batch_normal=[]
        self.pools=[]
        self.dropouts=[]
        self.layer_outputs=[]
        
        for i,(ft,kn,st) in enumerate(zip(filters,kernel_sizes,strides)):
            if kn<0:
                conv_function=tf.keras.layers.Conv2D
                conv_name_sub='_cv1d_l'
            else:
                conv_function=tf.keras.layers.Conv2DTranspose
                conv_name_sub='_decv1d_l'
            
            conv_name=conv_names[i] if len(conv_names)>i and len(conv_names[i])>0 else self.name+conv_name_sub+str(i+1) 
            pool_name=self.name+'_pool_l'+str(i+1)
            drop_name=self.name+'_drop_l'+str(i+1)
            
            # Check activation function name
            act = activations[i] if len(activations)>i and type(activations[i]) is type(tf.nn.leaky_relu) else None
            self.activation.append(act)
            
            # Check batch_normal
            if len(batch_normal) > i and batch_normal[i]==True:
                m,e=0.99,0.001
                if len(batch_norm_pars)>1:
                    if batch_norm_pars[0]>0 and batch_norm_pars[0]<1.:
                        m=batch_norm_pars[0]
                    if batch_norm_pars[1]>0 and batch_norm_pars[1]<1.:
                        e=batch_norm_pars[1]
                self.batch_normal.append(tf.keras.layers.BatchNormalization(momentum=m, epsilon=e, name=self.name+conv_name_sub+str(i+1)+'_bn'))
            else:
                self.batch_normal.append(None)
            
            # define conv
            self.convs.append(conv_function(
                filters=ft, kernel_size=(1,abs(kn)),strides=(1,st),
                kernel_initializer=kernel_initializer,
                padding='same',
                name=conv_name))
            self.conv_names.append(conv_name)
            
            # Check pool size for each layer: if pool size is small don't do max pool
            po = pool_sizes[i] if len(pool_sizes) > i else 0
            if po>1:
                po=int(po)
                self.pools.append(tf.keras.layers.AveragePooling2D(pool_size=(1,po),strides=(1,2),padding='same',name=pool_name))
            elif po<0:
                po=-int(po)
                self.pools.append(tf.keras.layers.MaxPool2D(pool_size=(1,po),strides=(1,2),padding='same',name=pool_name))
            else:
                self.pools.append([])
            
            # Check drop out value for each layer: if necessary (do>=1.), don't drop out
            do = drop_outs[i] if len(drop_outs) > i else 0
            if do<1.:
                self.dropouts.append(tf.keras.layers.Dropout(do,name=drop_name))
            else:
                self.dropouts.append([])
    
    def __call__(self, input_tensor, is_training=False, y_concat=None):
        with tf.compat.v1.variable_scope(self.name,reuse=tf.compat.v1.AUTO_REUSE):
            # expand input dim for conv1d to conv2d
            x=tf.expand_dims(input_tensor,1)
            # layers for convs
            for conv, pool, dropout, act, bn in zip (self.convs, self.pools, self.dropouts, self.activation, self.batch_normal):
                # Concatenate conditioning vector on feature map axis.
                if y_concat!=None and self.y_concat_ch==y_concat.get_shape()[-1]:
                    x_shapes=x.get_shape()
                    x=tf.concat([x, tf.reshape(y_concat, [self.batch_size, 1, 1, self.y_concat_ch])*
                                    tf.ones([self.batch_size, x_shapes[1], x_shapes[2], self.y_concat_ch])],-1)
                # 2d convolution
                x=conv(inputs=x)
                # batch normalization
                if bn is not None:
                    x=bn(x)
                tf.add_to_collection('DL', x)
                tf.add_to_collection('DL_ref', x)
                tf.add_to_collection('STD', x)
                # activation function
                if act is not None:
                    x=act(x)
                # pooling
                if pool!=[]:
                    x=pool(inputs=x)
                # dropout
                if dropout!=[]:
                    x = tf.compat.v1.where(is_training, dropout(inputs=x), x)
                # store each layer output
                self.layer_outputs.append(x)
                tf.add_to_collection('DL', x)
                tf.add_to_collection('DL_ref', x)
                tf.add_to_collection('STD', x)
            # squeeze outputs of conv2d to conv1d
            out=tf.squeeze(x,1)
            tf.add_to_collection('STD', out)
            return out
        
    def flatten(self, input_tensor, is_training=False, y_concat=None):
        return tf.keras.layers.Flatten()(self.__call__(input_tensor, is_training, y_concat))
    
    def get_output_number(self):
    # final output number if flatten
        return np.prod(self.output_shape)
```

```python
import numpy as np
import tensorflow as tf

class FcLayers():
    def __init__(self,
                 name='fc',
                 nodes=(),
                 trans_functions=(),
                 is_batch_norms=(),
                 do_rates=(),
                 y_concat_ch=0,
                 batch_norm_pars=(),):
        self.name=name
        self.nodes=nodes
        self.trans_funcs=trans_functions
        self.is_batch_norms=is_batch_norms
        self.do_rates=do_rates
        self.num_layers = 0 if len(nodes) < 2 else len(nodes)-2
        self.layer_names = []
        self.par_names=["kernel_fc_","bias_"]
        self.y_concat_ch=y_concat_ch
        self.batch_norm_pars=batch_norm_pars
        self.layer_outputs = []
        
        for i in range(self.num_layers+1):
            # Layer Name
            layer_name="_op" if i==self.num_layers else "_l"+str(i+1)
            layer_name=self.name+layer_name
            self.layer_names.append(layer_name)
    
    def __call__(self, input_tensor, is_training=False, y_concat=None):
        x=input_tensor
        for i in range(self.num_layers+1):
            is_batch_norm = self.is_batch_norms[i] if len(self.is_batch_norms) > i else False
            do_rate = self.do_rates[i] if len(self.do_rates) > i else 0.0
            trans_func = self.trans_funcs[i] if len(self.trans_funcs) > i else ''
            
            # Affine
            x = self.forward(x,self.nodes[i],self.nodes[i+1],self.layer_names[i],
                             trans_func,
                             is_training,
                             is_batch_norm,
                             do_rate,
                             y_concat)
        
        return x

    def trans_func(self,x,layer):
        return {
            'llu' : tf.nn.leaky_relu(layer),
            'elu' : tf.nn.elu(layer),
            'relu': tf.nn.relu(layer),
            'tanh': tf.nn.tanh(layer),
            'sigm': tf.nn.sigmoid(layer),
            'softmax': tf.nn.softmax(layer)
        }.get(x,layer)
    
    def forward(self, x, input_size, output_size, layer_name="",
              func_type=None,
              is_training=False,
              is_batch_norm=False,
              do_rate=0.0,
              y_concat=None):
        with tf.compat.v1.variable_scope(layer_name,reuse=tf.compat.v1.AUTO_REUSE):
            
            # TRAINABLE_VARIABLES
            W_name=self.par_names[0]+layer_name
            b_name=self.par_names[1]+layer_name
                
            if func_type=='sigm' or 'tanh':
                # Xavier
                c=1
            else:
                # He
                c=np.sqrt(2)
            
            if y_concat is not None and self.y_concat_ch==y_concat.get_shape()[-1]:
                input_size+=self.y_concat_ch
            
            st_dev=c/np.sqrt(input_size+output_size)
            init=tf.random.normal(shape=[int(input_size), int(output_size)],stddev=st_dev)
            
            W=tf.compat.v1.get_variable(name=W_name,initializer=init)
            b=tf.compat.v1.get_variable(name=b_name,shape=[output_size],initializer=tf.compat.v1.zeros_initializer)
            
            # concatenate y_concat
            if y_concat is not None and self.y_concat_ch==y_concat.get_shape()[-1]:
                x = tf.concat([x, y_concat], -1)
            
            #batch norm
            if is_batch_norm:
                m,e=0.99,0.001
                if len(self.batch_norm_pars)>1:
                    if self.batch_norm_pars[0]>0 and self.batch_norm_pars[0]<1.:
                        m=self.batch_norm_pars[0]
                    if self.batch_norm_pars[1]>0 and self.batch_norm_pars[1]<1.:
                        e=self.batch_norm_pars[1]
                x = tf.keras.layers.BatchNormalization(momentum=m, epsilon=e,name=layer_name+'_bn')(inputs=x)
            
            # multiply x and W
            mm = tf.matmul(x,W)
            
            # dropout if train
            if do_rate>0.0 and do_rate<1.0 and is_training is True:
                mm = tf.nn.dropout(mm,rate=1 - (do_rate))
            
            # add bias
            layer = tf.add(mm, b)
            self.layer_outputs = layer
            tf.add_to_collection('DL', self.layer_outputs)
            tf.add_to_collection('DL_ref', self.layer_outputs)
            tf.add_to_collection('STD', self.layer_outputs)

            out = self.trans_func(func_type,layer)
            tf.add_to_collection('DL', out)
            tf.add_to_collection('DL_ref', out)
            tf.add_to_collection('STD', out)
            return out
            
```

```python
import os
import pandas as pd
import random
import numpy as np
import itertools
import tensorflow as tf
import sys
import re
sys.path.insert(0,'../lib')

class NameSpacer:
    def __init__(self, **kwargs):
        self.__dict__.update(kwargs)

class Architecture:
    def __init__(self,
                 input_shape,
                 n_latent,
                 vae_encoder_cnn=None,
                 vae_encoder_mlp=None,
                 vae_decoder_mlp=None,
                 vae_decoder_cnn=None,
                 vae_decoder_cnn_input_shape=(0,),
                 sigma_x=1.,
                 beta=1.,
                 opt='adam',
                 loss_type='log',
                 loss_factor=1.0,
                 lr_start=0.001,
                 lr_decay_gstep=20000000000,
                 lr_decay_rate=0.98,
                 regularizer_scale=0.0,
                 regularizer_norm=1.0,
                 regularizer_base_num=10000.,
                 is_clip_out=False,
                 is_same_shape_inout=True,
                 output_shape=(),
                 x_aug_ch=-1,
                 y_aug_ch=-1,
                 yd_gamma=None,
                 lh_factor=1.,
                 name="VAE_CNN_Achitec"):
        self.input_shape=input_shape
        self.output_shape=input_shape if is_same_shape_inout or len(output_shape)==0 else output_shape
        self.x_aug_ch=x_aug_ch
        self.y_aug_ch=y_aug_ch
        if yd_gamma is None:
            self.yd_gamma=None
        else:
            self.yd_gamma=np.exp(-yd_gamma*np.arange(output_shape[0])) if yd_gamma>0 else np.exp(yd_gamma*np.flip(np.arange(output_shape[0])))
        self.lh_factor=lh_factor
        self.n_latent=n_latent
        self.vae_encoder_cnn=vae_encoder_cnn
        self.vae_encoder_mlp=vae_encoder_mlp
        self.vae_decoder_mlp=vae_decoder_mlp
        self.vae_decoder_cnn=vae_decoder_cnn
        self.vae_decoder_cnn_input_shape=vae_decoder_cnn_input_shape
        self.sigma_x=sigma_x
        self.beta=beta
        self.opt=opt
        self.loss_type=loss_type
        self.loss_factor=loss_factor
        
        self.starter_lr=lr_start
        self.lr_decay_gstep=lr_decay_gstep
        self.lr_decay_rate=lr_decay_rate
        
        self.regularizer_scale=regularizer_scale
        self.regularizer_norm=regularizer_norm
        self.regularizer_base_num=regularizer_base_num
        self.is_clip_out=is_clip_out
        self.name=name
        self.define_computation_graph()
        
        #Aliases
        self.ph = self.placeholders
        self.op = self.optimizers
        self.summ = self.summaries
        
    def define_computation_graph(self):
        # Reset graph
        tf.compat.v1.reset_default_graph()
        self.placeholders = NameSpacer(**self.define_placeholders())
        self.core_model = NameSpacer(**self.define_core_model())
        self.losses = NameSpacer(**self.define_losses())
        self.optimizers = NameSpacer(**self.define_optimizers())
        self.summaries = NameSpacer(**self.define_summaries())
    
    def define_placeholders(self):
        with tf.compat.v1.variable_scope("Placeholders"):
            # x_hat : inputs of VAE input CNN
            x_hat = tf.compat.v1.placeholder(dtype=tf.float32,
                               shape=(None,self.input_shape[0],self.input_shape[1]),
                               name="x_hat")
            tf.add_to_collection('DL', x_hat)
            x_ref = tf.compat.v1.placeholder(dtype=tf.float32,
                               shape=(None,self.input_shape[0],self.input_shape[1]),
                               name="x_ref")
            tf.add_to_collection('DL_ref', x_ref)
            x_aug = tf.compat.v1.placeholder(dtype=tf.float32,
                               shape=(None,abs(self.x_aug_ch)),
                               name="x_aug")
            # y : targets of VAE to be compared with reconstructed signal y
            y = tf.compat.v1.placeholder(dtype=tf.float32,
                               shape=(None,self.output_shape[0],self.output_shape[1]),
                               name="y")
            y_aug = tf.compat.v1.placeholder(dtype=tf.float32,
                               shape=(None,abs(self.y_aug_ch)),
                               name="y_aug")
            z_in = tf.compat.v1.placeholder(dtype=tf.float32,
                                  shape=(None,self.n_latent),
                                  name="z_in")
            is_train = tf.compat.v1.placeholder(dtype=tf.bool,name="is_train")
            acc_dev  = tf.compat.v1.placeholder(dtype=tf.float32,shape=None,name="acc_dev")
            loss_dev = tf.compat.v1.placeholder(dtype=tf.float32,shape=None,name="loss_dev")
            return({"x_hat": x_hat, "x_ref": x_ref, "x_aug": x_aug, "y": y, "y_aug": y_aug, "z_in": z_in, "is_train": is_train, "acc_dev": acc_dev, "loss_dev": loss_dev})
    
    def gaussian_encoder(self,x):
        if self.vae_encoder_cnn is not None:
            out=self.vae_encoder_cnn.flatten(x,self.placeholders.is_train)
            # tf.add_to_collection('DL', out)
            # tf.add_to_collection('DL_ref', out)
            if self.vae_encoder_mlp is not None:
                if self.x_aug_ch>0:
                    out=tf.concat([out,self.placeholders.x_aug],1)
                # tf.add_to_collection('STD', out)
                out=self.vae_encoder_mlp(out,self.placeholders.is_train)
        else:
            x=tf.reshape(x,[-1,self.input_shape[0]*self.input_shape[1]])
            if self.x_aug_ch>0:
                x=tf.concat([x,self.placeholders.x_aug],1)
            out=self.vae_encoder_mlp(x,self.placeholders.is_train)
        mean=out[:,:self.n_latent]
        stddev=1e-6+tf.nn.softplus(out[:,self.n_latent:])
        return mean, stddev
    
    def gaussian_decoder(self,z):
        if self.y_aug_ch>0:
            z=tf.concat([z,self.placeholders.y_aug],1)
        tf.add_to_collection('STD', z)
        out=self.vae_decoder_mlp(z,self.placeholders.is_train)
        # tf.add_to_collection('DL', out)
        # tf.add_to_collection('DL_ref', out)
        if self.vae_decoder_cnn is not None:
            out=tf.reshape(out,[-1,self.vae_decoder_cnn_input_shape[0],
                                   self.vae_decoder_cnn_input_shape[1]])
            out=self.vae_decoder_cnn(out,self.placeholders.is_train)
        else:
            out=tf.reshape(out,[-1,self.output_shape[0],self.output_shape[1]])
        return tf.clip_by_value(out, 1e-8, 1-1e-8) if self.is_clip_out==True else out
    
    def define_core_model(self):
        with tf.compat.v1.variable_scope("Core_Model"):
            # encoding
            mu, sigma = self.gaussian_encoder(self.placeholders.x_hat)
            mu_ref, sigma_ref = self.gaussian_encoder(self.placeholders.x_ref)            
            # sampling by re-parameterization technique while training
            z=mu+sigma*tf.random.normal(tf.shape(input=mu),0,0,dtype=tf.float32)
            z_ref=mu_ref+sigma_ref*tf.random.normal(tf.shape(input=mu_ref),0,0,dtype=tf.float32)
            # decoding
            y_pred=self.gaussian_decoder(z)
            y_pred_ref=self.gaussian_decoder(z_ref)
            #y_fixed = self.gaussian_decoder(mu)
            # decoding for generation
            #y_gen=self.gaussian_decoder(self.placeholders.z_in)
            
            return({"mu":mu,"sigma":sigma,"y_pred":y_pred,"z": z,"mu_ref":mu_ref,"sigma_ref":sigma_ref,"y_pred_ref":y_pred_ref,"z_ref":z_ref})#,"y_gen":y_gen, "y_fixed": y_fixed})

    def define_losses(self):
        with tf.compat.v1.variable_scope("Losses"):
            mu=self.core_model.mu
            sigma=self.core_model.sigma
            y=tf.reshape(self.placeholders.y,[-1,self.output_shape[0]*self.output_shape[1]])
            y_pred=tf.reshape(self.core_model.y_pred,  [-1,self.output_shape[0]*self.output_shape[1]])
            
            # marginal likelihood
            if self.loss_type=='log':
                marginal_likelihood=tf.reduce_sum(input_tensor=y*tf.math.log(y_pred)+(1-y)*tf.math.log(1-y_pred), axis=1)
            else:
                eps=1e-5
                y_clip=tf.clip_by_value(y,eps,1-eps)
                y_pred_clip=tf.clip_by_value(y_pred,eps,1-eps)
                marginal_likelihood=tf.reduce_sum(input_tensor=self.lh_factor*y_clip*tf.math.log(self.lh_factor*y_pred_clip)+
                                                 (self.lh_factor-self.lh_factor*y_clip)*tf.math.log(self.lh_factor-self.lh_factor*y_pred_clip),
                                                  axis=1)
            
            # loss of data reconstruction
            if self.loss_type=='mae':
                if self.yd_gamma is None:
                    loss_reconst=-tf.reduce_sum(input_tensor=tf.abs(y-y_pred), axis=1)
                    loss_reconst_wo_gamma=loss_reconst
                else:
                    loss_reconst=-tf.reduce_sum(input_tensor=self.yd_gamma*tf.abs(y-y_pred), axis=1)
                    loss_reconst_wo_gamma=-tf.reduce_sum(input_tensor=tf.abs(y-y_pred), axis=1)
            elif self.loss_type=='mse':
                if self.yd_gamma is None:
                    loss_reconst=-tf.reduce_sum(input_tensor=tf.square(y[:15]-y_pred[:15]), axis=1)
                    loss_reconst_wo_gamma=loss_reconst
                else:
                    loss_reconst=-tf.reduce_sum(input_tensor=self.yd_gamma*tf.square(y-y_pred), axis=1)
                    loss_reconst_wo_gamma=-tf.reduce_sum(input_tensor=tf.square(y-y_pred), axis=1)
            else:
                loss_reconst=marginal_likelihood
                loss_reconst_wo_gamma=loss_reconst
            
            # loss of divergence
            if self.sigma_x==1.:
                KL_divergence=0.5*tf.reduce_sum(input_tensor=tf.square(mu)+tf.square(sigma)-tf.math.log(1e-8+tf.square(sigma))-1,axis=1)
            else:
                sqr_sigx=tf.square(self.sigma_x)+1e-8
                KL_divergence=0.5*tf.reduce_sum(input_tensor=(tf.square(mu)+tf.square(sigma))/sqr_sigx+tf.math.log(sqr_sigx/(1e-8+tf.square(sigma)))-1,axis=1)

            # average
            if self.loss_type=='log':
                marginal_likelihood=tf.reduce_mean(input_tensor=marginal_likelihood)
            loss_reconst_batch=loss_reconst
            loss_reconst=tf.reduce_mean(input_tensor=loss_reconst)
            loss_reconst_wo_gamma=tf.reduce_mean(input_tensor=loss_reconst_wo_gamma)
            KL_divergence=tf.reduce_mean(input_tensor=KL_divergence)

            if self.loss_type!='log':
                loss=self.beta*KL_divergence-self.loss_factor*loss_reconst
            else:
                loss=self.beta*KL_divergence-loss_reconst
            
            # loss with regularizer
            if self.regularizer_scale>0.0:
                vars=tf.trainable_variables()
                lp_loss=tf.add_n([tf.norm(tensor=v,ord=self.regularizer_norm) for v in vars if 'bias' not in v.name])
                var_num_total=float(int(np.sum([np.prod(v.shape) for v in vars if 'bias' not in v.name])))
                
                loss_reg=loss+self.regularizer_scale*lp_loss*self.regularizer_base_num/var_num_total
            else:
                loss_reg=loss
            
            return({"loss_reg": loss_reg, "loss": loss,
                    "loss_reconst": loss_reconst, "loss_reconst_wo_gamma": loss_reconst_wo_gamma, "loss_reconst_batch": loss_reconst_batch,
                    "marginal_likelihood": marginal_likelihood, "KL_divergence": KL_divergence})
    
    # optimize loss
    def define_optimizers(self):
        with tf.compat.v1.variable_scope("Optimization"):
            def opt_type(x):
                return {
                    'adadelta':tf.compat.v1.train.AdadeltaOptimizer,
                    'adagrad':tf.compat.v1.train.AdagradOptimizer,
                    'adagradda':tf.compat.v1.train.AdagradDAOptimizer,
                    'adam':tf.compat.v1.train.AdamOptimizer,
                    'momentum':tf.compat.v1.train.MomentumOptimizer,
                    'rmsprop':tf.compat.v1.train.RMSPropOptimizer
                }.get(x,tf.compat.v1.train.AdamOptimizer)
            
            train_vars = tf.compat.v1.get_collection(tf.compat.v1.GraphKeys.TRAINABLE_VARIABLES, scope='Core_Model')
            self.global_step=tf.Variable(0,trainable=False)
            self.lr=tf.compat.v1.train.exponential_decay(self.starter_lr,
                                               self.global_step,
                                               self.lr_decay_gstep,
                                               self.lr_decay_rate,
                                               staircase=True)
            self.optimizer=opt_type(self.opt)(learning_rate=0.0005, beta1=0.9, beta2=0.999, epsilon=1e-08, name='Adam')#learning_rate=self.lr)
            op = self.optimizer.minimize(self.losses.loss_reg, var_list=train_vars, global_step=None)#self.global_step)
            return({"op": op})
    
    def define_summaries(self):
        with tf.compat.v1.variable_scope("Summaries"):
            loss = self.losses.loss
            return({"loss": loss})

```

```python
import tensorflow as tf
import time



bSize = 1000
dim_z = 128
start = time.time()

loss_type='mse'

# cnn of encoder
E_cnn_input_shape = X_e.shape[1:]
E_cnn_kernels =      (16,32,64,64,32)
E_cnn_kernels_length = (5,)*len(E_cnn_kernels)
E_cnn_strides =        (1,2,2,2,2,)
E_cnn_pool_sizes =     (1,)*len(E_cnn_kernels)
E_cnn_drop_outs =      (1,)*len(E_cnn_kernels)
E_cnn_acts = (tf.nn.leaky_relu,)*len(E_cnn_kernels)
E_cnn_batchnorm=(False,)*len(E_cnn_kernels)

E_cnn = Conv1dLayers(name='E_cnn', batch_size=bSize, 
                     input_shape=E_cnn_input_shape, filters=E_cnn_kernels,
                     kernel_sizes=E_cnn_kernels_length,
                     kernel_initializer='glorot_uniform',
                     strides=E_cnn_strides, pool_sizes=E_cnn_pool_sizes,
                     drop_outs=E_cnn_drop_outs, activations=E_cnn_acts,
                     batch_normal=E_cnn_batchnorm)

# mlp of encoder
E_mlp_nodes=np.array([E_cnn.get_output_number(),
                      dim_z*2])
E_mlp_nn_func_type = ('relu',)*(max(len(E_mlp_nodes)-2,0))+('',)*(len(E_mlp_nodes)-1)
E_mlp_batch_norm=('',)*(len(E_mlp_nodes)-1)
E_mlp_keep_prob=(0.5,)*(len(E_mlp_nodes)-1)

E_mlp=FcLayers(
    'E_mlp',
     E_mlp_nodes,
     E_mlp_nn_func_type,
     E_mlp_batch_norm,
     E_mlp_keep_prob)

# cnn of decoder
D_cnn_input_shape = (4,32) # the same with E_cnn final output shape
D_cnn_output_shape = (y_d.shape[-1],1)
D_cnn_kernels =      (32,128,128,1)
D_cnn_kernels_length = (5,5,5,3,)
D_cnn_strides =        (1,2,2,1,)
D_cnn_pool_sizes =     (1,)*len(D_cnn_kernels)
D_cnn_drop_outs =      (0.5,)*len(D_cnn_kernels)
D_cnn_acts = (tf.nn.leaky_relu,)*(len(D_cnn_kernels)-1)+(tf.nn.relu,)
D_cnn_batchnorm=(False,)*len(D_cnn_kernels)

D_cnn = Conv1dTransLayers(
    name='D_decnn',
    batch_size=bSize,
    input_shape=D_cnn_input_shape,
    output_shape=D_cnn_output_shape,
    filters=D_cnn_kernels,
    kernel_sizes=D_cnn_kernels_length,
    kernel_initializer='glorot_uniform',
    strides=D_cnn_strides,
    pool_sizes=D_cnn_pool_sizes,
    drop_outs=D_cnn_drop_outs,
    activations=D_cnn_acts,
    batch_normal=D_cnn_batchnorm)

# mlp of decoder
D_mlp_nodes=np.array([dim_z+X_d.shape[-1]+1,
                      np.prod(D_cnn_input_shape)])
D_mlp_nn_func_type = ('llu','',)
D_mlp_batch_norm=('',)*(len(D_mlp_nodes)-1)
D_mlp_keep_prob=(0.8,)*(len(D_mlp_nodes)-1)

D_mlp=FcLayers(
    'D_mlp',
     D_mlp_nodes,
     D_mlp_nn_func_type,
     D_mlp_batch_norm,
     D_mlp_keep_prob)

# input and output nodes for architecture
net = Architecture(input_shape=X_e.shape[1:],
                   n_latent=dim_z,
                   vae_encoder_cnn=E_cnn,
                   vae_encoder_mlp=E_mlp,
                   vae_decoder_cnn=D_cnn,
                   vae_decoder_mlp=D_mlp,
                   vae_decoder_cnn_input_shape=D_cnn_input_shape,
                   beta=1,
                   loss_type=loss_type,
                   loss_factor=1000.,
                   lr_start=0.0005,
                   regularizer_scale=2.,
                   regularizer_norm=2.,
                   is_clip_out=False,
                   is_same_shape_inout=False,
                   output_shape=D_cnn_output_shape,
                   y_aug_ch=X_d.shape[-1]+1)

saver = tf.train.Saver()

config = tf.ConfigProto()
config.gpu_options.allow_growth = True
sess = tf.InteractiveSession(config=config)

sess.run(tf.global_variables_initializer())

# with tf.Session() as sess:
# saver.restore(sess,tf.train.latest_checkpoint(log_dir))
saver.restore(sess,'./ckpt/model.ckpt-999')


print("time model making:", time.time() - start)
```

```python
def lrp(x_idx, num_rlt):

    y_aug=np.concatenate((X_d_test[x_idx:x_idx+1],np.expand_dims(X_e_test[x_idx:x_idx+1,-1,-1],-1)),axis=1)
    feed_dict = {net.ph.x_hat: X_e_test[x_idx:x_idx+1],
                net.ph.y: np.expand_dims(y_d_test,-1), 
                net.ph.y_aug: y_aug,
                net.ph.is_train: False,
                net.ph.z_in: np.zeros((1,128))}

    nodes = tf.get_collection('STD')
    del nodes[36:]
    del nodes[12:24]

    n23 = nodes[23][:,num_rlt:num_rlt+1,:]
    n12 = nodes[12]
    # n12_e = n12[:,:128]
    # n12_d = n12[:,128:]
    R23 = n23
    R12 = n12 * tf.gradients(R23, n12)[0]
    R12_e = n12[:,:128] * tf.gradients(R23, n12)[0][:,:128]
    R12_d = y_aug * tf.gradients(R23, n12)[0][:,128:]
    R0 = net.ph.x_hat * tf.gradients(R23, net.ph.x_hat)[0]


    r23 = sess.run(R23, feed_dict=feed_dict)
    r12_e = sess.run(R12_e / tf.reduce_sum(R12) * tf.reduce_sum(R23), feed_dict=feed_dict)
    r12_d = sess.run(R12_d / tf.reduce_sum(R12) * tf.reduce_sum(R23), feed_dict=feed_dict)
    r0 = sess.run(R0 / tf.reduce_sum(R0) * tf.reduce_sum(r12_e), feed_dict=feed_dict)
    return r0, r12_e, r12_d, r23
```

```python
def lime(x_idx, num_rlt):
    import pandas as pd
    from functools import partial
    from sklearn.utils import check_random_state
    from sklearn import preprocessing, metrics
    from sklearn.linear_model import Ridge
    kernel_multiplier = 0.3
    training_data = Xe_Xd
    kernel_width = np.sqrt(training_data.shape[1]) * kernel_multiplier
    # kernel_width = float(kernel_width)
    def kernel(d, kernel_width):
        return np.sqrt(np.exp(-(d ** 2) / kernel_width ** 2))
    kernel_fn = partial(kernel, kernel_width=kernel_width)
    # scaling
    scaler = preprocessing.StandardScaler(with_mean=False)
    scaler.fit(training_data)
    y_aug=np.concatenate((X_d_test[x_idx:x_idx+1],np.expand_dims(X_e_test[x_idx:x_idx+1,-1,-1],-1)),axis=1)
    random_state = None; random_state = check_random_state(random_state)
    num_samples = 150000 ; alpha = 1 ; data_row = Xe_Xd_test[x_idx:x_idx+1]
    num_cols = data_row.shape[1]
    data = random_state.normal(0, 1, num_samples * num_cols).reshape(num_samples, num_cols)
    data = data * scaler.scale_ + scaler.mean_
    data[0] = data_row.copy()
    scaled_data = (data - scaler.mean_) / scaler.scale_
    distances = metrics.pairwise_distances(scaled_data, scaled_data[0].reshape(1, -1), metric='euclidean').ravel()

    def limefunc(data): 
        Xe_flat = data[:,:128*11]
        Xd = data[:,128*11:]
        Xe = Xe_flat.reshape(-1,128,11)#.squeeze()
        y_aug=np.concatenate((Xd,np.expand_dims(Xe[:,-1,-1],-1)),axis=1)
        y_pred_test = sess.run([net.core_model.y_pred,], feed_dict={net.ph.x_hat: Xe, net.ph.y_aug: y_aug, net.ph.is_train: False})
        return np.array(y_pred_test).squeeze()[:,num_rlt]

    neighborhood_labels = limefunc(data)
    neighborhood_data = scaled_data
    weights = kernel_fn(distances)

    easy_model = Ridge(alpha=alpha, fit_intercept=False, random_state=random_state)
    easy_model.fit(neighborhood_data, neighborhood_labels, sample_weight=weights)
    prediction_score = easy_model.score(neighborhood_data, neighborhood_labels, sample_weight=weights)
    local_pred = easy_model.predict(neighborhood_data[0].reshape(1, -1))
    intercept = easy_model.intercept_
    coeff = easy_model.coef_
    expco = easy_model.coef_ * (data_row - scaler.mean_) / scaler.scale_
#     print(coeff)
#     print(expco)
#     print(prediction_score, intercept)
#     print(np.sum(expco) + intercept)
    print('num_rlt = ', num_rlt)
    return coeff, expco, prediction_score, intercept


    # with open('lime_9132.pickle','wb') as f:
#     pickle5.dump([EXPCO, COEFF, SCORE, INTERCEPT], f)
```
