```python
import tensorflow as tf
import tensorflow.keras as tk
import numpy as np
import matplotlib.pyplot as plt
```

### Parameters

```python
past_stepsize = 1
past_windowsize = 20
pred_stepsize = 10
pred_windowsize = 10
```

### Functions

```python
def pos_encoder(pos, freq=400, d=4):
    pos_enc = np.zeros([pos.shape[0], d], dtype=np.float32)
    for i in range(d):
        if i % 2 == 0:
            pos_enc[:, i] = np.sin(pos / (freq ** (i / d)))
        else:
            pos_enc[:, i] = np.cos(pos / (freq ** (i / d)))
    return pos_enc

def data_loader(file_prefix, phase):
    # past data column: ['pv', 'aglSteering', 'prBrakeMasterCylinder', 'distObjectAcc', 'spdRelObjectAcc', 'vsTcu']
    data = np.load(file_prefix + 'dataset_past_' + str(phase) + '.npy', mmap_mode='r')
    data = data.astype('float32')
    past = data[:,  -past_stepsize*(past_windowsize-1)-1::past_stepsize]
    past = past.astype('float32')
    map_data = np.load(file_prefix + 'dataset_map_' + str(phase) + '.npy')
    map_data = map_data.astype('float32')
    pred = np.load(file_prefix + 'dataset_pred_' + str(phase) + '.npy')
    if pred_stepsize == 10:
        pred = pred[:, 1::2]
    pred = pred.astype('float32')
    dist = np.load(file_prefix + 'dataset_dist_' + str(phase) + '.npy')

    return past, map_data, pred, dist

    return past, map_data, pred, dist
```

### Load Trained Model

```python
results_path = './results/att_fwpos_20220602-150355/'
model_path = results_path+'model_pb/'
model = tk.models.load_model(model_path)
```

### Load data for Test

```python
file_prefix = '../data/hatci_clinic/1st_TD_matched/dataset_1/'
data_past, data_map, data_pred, data_dist = data_loader(file_prefix, 'test')

# data scale
past_max = np.load(results_path+'evaluations/past_max.npy')
scaled_data_past = data_past/past_max
map_max = np.load(results_path+'evaluations/map_max.npy')
scaled_data_map = data_map/map_max
pred_max = np.load(results_path+'evaluations/pred_max.npy')
scaled_data_pred = data_pred/pred_max
```

### Visualize dataset

```python
plt.figure(figsize=(15,5))
plt.plot(data_past[:11000,0,-1:])
plt.ylabel('Speed (Kph)')
plt.xlabel('Time (0.1 sec)')
plt.grid()
```

### 1. Model analysis w.r.t a sample

```python
# Parameters
idx = 8330 # Sample index
output_time_steps=10
pos_enc_past = tf.tile(tf.expand_dims(pos_encoder(np.arange(10), 40), axis=0), [1, 1, 1])
pos_enc_fwd = tf.tile(tf.expand_dims(pos_encoder(np.arange(100)), axis=0), [1, 1, 1])

# Data Sampling
x_past = scaled_data_past[idx:idx+1]
x_forward = scaled_data_map[idx:idx+1]
y = scaled_data_pred[idx:idx+1]
print(f'Past data shape: {x_past.shape}')
print(f'Forward data shape: {x_forward.shape}')
print(f'Prediction target data shape: {y.shape}')
```

#### 1.1 Inference

```python
output = model(x_past, x_forward, pos_enc_past, pos_enc_fwd)
```

```python
# Error evaluation
error = pred_max * (y-output)
mse = np.mean(np.square(error), axis=0)
mae = np.mean(np.abs(error), axis=0)

np.set_printoptions(precision=4, suppress=True)
print("mse")
print(mse)
print("mean = ", np.mean(mse))
print("mae")
print(mae)
print("mean = ", np.mean(mae))

# Results Visualize
x1 = np.arange(1, 21)*0.1 - 2
x2 = np.arange(1, 11)
plt.figure(figsize=(15,5))
plt.plot(x1, (x_past*past_max)[0, :, -1], '.-', label='past')
plt.plot(x2, y[0]*pred_max[0], '.-', label='truth')
plt.plot(x2, output[0]*pred_max, '.-', label='pred')
plt.xlabel('time (sec)')
plt.ylabel('speed (kph)')
plt.legend()
plt.grid()
```

이제 아래와 같이 Dense layer를 두개 거쳐 Attention에 사용될 전방정보를 뽑아냅니다. Dense layer는 입력값의 마지막 dimension에 대해서만 적용됩니다.

```python
feature = tf.concat([x_forward, pos_enc_fwd], axis=-1)
feature = model.spatial_att.relu(model.spatial_att.dense2(model.spatial_att.relu(model.spatial_att.dense1(feature))))

plt.figure(figsize=(15,5))
plt.imshow(tf.transpose(feature, [0, 2, 1])[0])
```

앞서 만든 Query와 전방정보를 가지고 attention mechanism을 적용합니다.

```python
v1 = model.spatial_att.u_e(feature)
v2 = model.spatial_att.w_e(query)

# visualize v1
plt.figure(figsize=(15,5))
plt.imshow(tf.transpose(v1, [0, 2, 1])[0])
# plt.colorbar()
plt.title('v1 in attention')

plt.figure(figsize=(15,5))
for i in range(v1.shape[-1]):
    plt.plot(v1[0, :, i], label=str(i))
plt.xlim([0, 100])
plt.title('v1 in attention')
plt.xlabel('distance (m/5)')
# plt.legend()
plt.grid()

# visualize v2
plt.figure(figsize=(15,5))
for i in range(v2.shape[1]):
    plt.plot(v2[0, i], label='Query T+'+str(i+1))
plt.title('v2 in attention')
plt.legend()
plt.grid()
```

위의 Figure에서 v1의 칼럼 하나가 attention mechanism의 x^i 라고 생각하면 됩니다.

이제 attention weight alpha를 계산합니다.

```python
v2 = tf.expand_dims(v2, axis=2)  # shape: [batch, self.output_time_steps, 1, units]
e_t = tf.squeeze(model.spatial_att.v_e(model.spatial_att.tanh(v1 + v2)), axis=-1)  # shape: [batch, self.output_time_steps, feature_length]
alpha_t = model.spatial_att.softmax(e_t)  # shape: [batch, self.output_time_steps, feature_length]
```

```python
# visualize e and alpha
plt.figure(figsize=(15,5))
for i in range(e_t.shape[1]):
    plt.plot(e_t[0, i], label='T+'+str(i+1))

plt.xlim([0, 100])
plt.xlabel('distance')
plt.title('e')
plt.legend()
plt.grid()

plt.figure(figsize=(15,5))
plt.imshow(e_t[0])
plt.title('e')

plt.figure(figsize=(15,5))
for i in range(alpha_t.shape[1]):
    plt.plot(alpha_t[0, i], label='T+'+str(i+1))

plt.xlim([0, 100])
plt.xlabel('distance')
plt.title('alpha')
plt.legend()
plt.grid()

plt.figure(figsize=(15,5))
plt.imshow(alpha_t[0])
plt.title('alpha')
```

다음으로는 계산된 alpha값을 가지고 attention된 전방정보를 계산합니다.

위와 같이 따로 연산하지 않고, Attention layer를 호출해 한번에 attention weight를 계산할 수도 있습니다.

```python
alpha = model.spatial_att(query, tf.expand_dims(tf.concat([x_forward, pos_enc_fwd],axis=-1), axis=1))
```

```python
# visualize
plt.figure(figsize=(15,5))
for i in range(alpha.shape[1]):
    plt.plot(alpha[0, i], label='T+'+str(i+1))
plt.xlim([0, 100])
plt.xlabel('distance')
plt.title('alpha')
plt.legend()
plt.grid()

plt.figure(figsize=(15,5))
plt.imshow(alpha[0])
plt.title('alpha')
```

```python
att_feature = tf.expand_dims(alpha_t, axis=3) * x_forward
```

```python
# visualize attention feature
plt.figure(figsize=(14.5,3))
plt.plot(x_forward[0], '.-')
plt.legend(['radius', 'slope', 'property', 'type'])
plt.xlim([0, 100])
plt.ylim([-1, 1])
plt.xlabel('ditance')
plt.ylabel('normalized value')
plt.title('Forward Information')
plt.grid()

plt.figure(figsize=(15,5))
plt.imshow(tf.transpose(x_forward[0]))
plt.title('raw forward information')
for i in range(att_feature.shape[1]):
    plt.figure(figsize=(15,5))
    plt.imshow(tf.transpose(att_feature[0, i]))
    plt.title(f'attention feature for T+{i+1}')
```

attention이 적용된 feature를 `enc_forward`와 dense layer를 통해 `code_forward`를 추출합니다.

```python
code_forward = model.relu(model.dense3(tf.reshape(model.enc_forward(att_feature), [-1, output_time_steps, 94 * 32])))
print(f'code forward shape: {code_forward.shape}')
```

##### 1.2.3 Predict future speed

Query와 code_forward를 concatenate해 최종 속도를 예측합니다.

```python
code = tf.concat([query, code_forward], axis=-1)  # shape: [-1, self.output_time_steps, k+m]
x = model.relu(model.dense4(code))  # shape: [-1, self.output_time_steps, 1]
output = model.tanh(model.dense5(x)) + x_past[:, -1:, -1:]
```

```python
# visualize results
plt.figure(figsize=(15,5))
plt.plot(x1, (x_past*past_max)[0, :, -1], '.-', label='past')
plt.plot(x2, y[0]*pred_max[0], '.-', label='truth')
plt.plot(x2, output[0]*pred_max, '.-', label='pred')
plt.xlabel('time (sec)')
plt.ylabel('speed (kph)')
plt.legend()
plt.grid()
```

#### 1.3 Attention weight w.r.t. travel distance

다음으로는 이동 거리에 따른 attention weight의 변화에 대해 살펴보겠습니다. 먼저, 실제 속도와 예측 속도에 기반해 이동거리를 계산합니다.

```python
dis_pred = []
dis_true = []
for i in range(10):
    if i == 0:
        dis_pred.append(((x_past*past_max)[0,-1,-1]+output[0,i]*pred_max)*1000/3600*0.5)
        dis_true.append(((x_past*past_max)[0,-1,-1]+y[0,i]*pred_max)*1000/3600*0.5)
    else:
        dis_pred.append((output[0,i-1]+output[0,i])*pred_max*1000/3600*0.5)
        dis_true.append((y[0,i-1]+y[0,i])*pred_max*1000/3600*0.5)
        
plt.figure(figsize=(15,5))
plt.plot(x2, np.cumsum(dis_pred), '.-', label='pred')
plt.plot(x2, np.cumsum(dis_true), '.-', label='truth')
plt.ylabel('travel distance (m)')
plt.xlabel('time (sec)')
plt.title('Travel distance w.r.t. speed')
plt.legend()
plt.grid()
```

```python
plt.figure(figsize=(15,5))
plt.plot(xf, x_forward[0])
plt.grid()
plt.legend(['radius', 'slope', 'property', 'type'])
plt.title('Forward information')
plt.xlabel('distance (m)')
plt.ylabel('normalized value')

plt.figure(figsize=(15,5))
for i in range(output_time_steps):
    plt.plot(xf, alpha_t[0,i], label=f'T+{i+1}')
    plt.axvline(np.cumsum(dis_pred)[i], linestyle='--', color='tab:gray')
    plt.axvline(np.cumsum(dis_true)[i], linestyle='--', color='tab:red')
plt.title('Attention weight and travel distance')
plt.xlabel('distance (m)')
plt.legend()
plt.grid()

plt.figure(figsize=(15,2))
plt.imshow(alpha_t[0])
plt.colorbar()
plt.title(f'spatial attention weight')
```
